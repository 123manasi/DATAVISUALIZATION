```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.stripplot(x='year',y='duration_ms',data=df,jitter='0.2',palette='muted')
plt.figure(figsize=(15,7))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.jointplot(x='year',y='instrumentalness',data=df,kind='reg')
plt.figure(figsize=(15,7))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.kdeplot(x='speechiness',hue='year',data=df)
plt.figure(figsize=(15,7))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x='year',y='loudness',data=df)
plt.figure(figsize=(15,7))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x='year',y='tempo',data=df)
plt.figure(figsize=(15,7))
```

```python
import seaborn as sns
sns.lineplot(x='key',y='instrumentalness',data=df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.swarmplot(x='year',y='energy',data=df,size=1.9,palette='dark')
plt.title('Apporva')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(x='year',data=df,kde=True)
plt.title('Apporva')
plt.show()
```
